```python
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mult(a, b):
    return a * b

def div(a, b):
    return a / b

def rem(a, b):
    return a % b

print("Simple Calculator")

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
print("5. Remainder")

op = input("Choose an option: ")

if op == "1":
    ans = add(a, b)
    print(ans)
elif op == "2":
    ans = sub(a, b)
    print(ans)
elif op == "3":
    ans = mult(a, b)
    print(ans)
elif op == "4":
    ans = div(a, b)
    print(ans)
elif op == "5":
    ans = rem(a, b)
    print(ans)
else:
    print("Wrong choice")5
```

```python
def tst8(a):
  l=[]
  for i in a :
    if type(i) == list:
      for j in i:
        l.append(j)
    else:
      if type(i) == int or type(i) == float:
        l.append(i)
  return l
```

```python
def test_fib(n):
  a,b = 0,1
  for i in range(n):
    yield a
    a,b = b ,a+b
```

```python
fib = test_fib(10)
for value in fib:
    print(value)
```

```python
def count_test(n):
  count = 1
  while count <= n :
    yield count
    count = count +1
```
